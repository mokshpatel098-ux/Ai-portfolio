Creat an array an called marks, and print the first mark, and print the last mark ,print marks from index 2 to 4, and print every second value

```python
marks = [14, 34, 56, 89, 45, 35, 67, 56, 54, 43]
print("First mark:", marks[0])
print("Last mark:", marks[-1])
print("Marks from index 2 to 4:", marks[2:5])
print("Every second value:", marks[::2])
```
